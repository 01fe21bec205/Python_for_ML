### *LINEAR REGRESSION BY GIVING X AND Y VALUES*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,5,13,25,33,46]).reshape((-1,1))
y=np.array([3,6,15,12,21,18])
```

```python
plt.scatter(x,y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
SLR=LinearRegression()
```

```python
SLR.fit(x,y)
```

```python
pred=SLR.predict(x)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='Orange',marker='*')
print("MSE:",mse)
print("intercept:",SLR.intercept_)
print("slope:",SLR.coef_)
```

### *LINEAR REGRESSION BY RANDOMLY GENERATING X AND Y VALUES*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x=[]
y=[]
for i in range(6):
  xx=random.randint(1,9)
  x.append(xx)
  yy=random.randint(1,9)
  y.append(yy)
w=np.array(x)
x=w.reshape((-1,1))
print(x)
print(y)
```

```python
plt.scatter(x,y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
SLR=LinearRegression()
```

```python
SLR.fit(x,y)
```

```python
pred=SLR.predict(x)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='Green',marker='*')
print("MSE:",mse)
print("intercept:",SLR.intercept_)
print("slope:",SLR.coef_)
```
