## ***3.	Create a game (like lottery/rock-paper-scissors) and write a python code to record how many times the computer won and the user won in an excel file. Plot the graph showing the success rate of user and computer with different colours.***

CODE : Consider a DICE ROLLING GAME where we generate random number from 1 to 6
***

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from openpyxl import load_workbook, Workbook

u = np.zeros(10)
c = np.zeros(10)

for i in range(10):
    computer = random.randint(1, 6)
    user = int(input("\nEnter your dice number: "))
    print("Computer's dice number:", computer)
    if computer > user:
        c[i] = 1
        u[i] = 0
        print("Computer scored")
    elif computer < user:
        c[i] = 0
        u[i] = 1
        print("User scored")

print("Computer scores:", c)
print("User scores:", u)
try:
    workbook = load_workbook("dice_rolling_results.xlsx")
except FileNotFoundError:
    workbook = Workbook()
    workbook.save("dice_rolling_results.xlsx")

sheet = workbook.active
df = pd.DataFrame([c, u], columns=range(1, 11), index=['Computer', 'User'])

for index, row in df.iterrows():
    sheet.append(row.tolist())

workbook.save("dice_rolling_results.xlsx")

mean = df.mean(axis=1)
print("Mean scores:", mean)

plt.pie(mean, labels=['Computer', 'User'])
plt.show()

```
