### **Lottery game using random module**

```python
import random
x=0
y=0
for z in range(1,10):
  userinput=int(input("enter number between 1,9 = "))
  computer=random.randint(1,9)
  print("computer num = " , computer)
  print("user number = " , userinput)

  if computer>userinput:
   print("computer won")
   x=x+1
  elif computer==userinput:
   print("game tie")
  else:
   print("user won")
   y=y+1

if(x>y):
   print("computer")
elif(x==y):
   print("tie")
else:
  print("user")
```

### **Rock, paper, scissor game using random module**

```python
import random
x=0
y=0
for z in range(1,10):
  gamelist=['rock','paper','scissor']
  u=input("enter rock or paper or scissor = ")
  c=random.choice(gamelist)
  print("computer = " , c)
  print("user = " , u)

  if ((c=="rock" and u=="scissor") or (c=="paper" and u=="rock" ) or (c=="scissor" and u=="paper")):
   print("computer won")
   x=x+1
  elif ((c=="rock" and u=="rock") or (c=="paper" and u=="paper" ) or (c=="scissor" and u=="scissor")):
   print("game tie")
  elif((u=="rock" and c=="scissor") or (u=="paper" and c=="rock" ) or (u=="scissor" and c=="paper")):
   print("user won")
   y=y+1


if(x>y):
   print("computer")
elif(x==y):
   print("tie")
else:
  print("user")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100)
y=4+2*np.sin(2*x)

fig, a=plt.subplots()
a.plot(x,y,linewidth=2.0)
a.set(xlim=(0,8), xticks=np.arange(1,8),
      ylim=(0,8), yticks=np.arange(1,8))
plt.grid()
plt.show()
```
